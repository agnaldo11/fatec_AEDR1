```python
#Peça dois números ao usuário e mostre a soma deles.
num1 = float(input('Digite o primeiro número: '))
num2 = float(input('Digite o segundo número: '))

soma = num1 + num2
print(f'Soma dos dois numeros são:{soma:.2f}') 
           
```

```python
# Próximo exercício problema 2 — Par ou Ímpar
num = int(input('Digite o primeiro número: '))

if num % 2 == 0:
    print(f'Número é par:{num}') 
else: 
    print(f'Número impar:{num}')
```

```python
#Peça 3 notas, calcule a média e mostre:
nota1 = float(input('Digite a primeira nota: '))
nota2 = float(input('Digite a segunda nota: '))
nota3 = float(input('Dgiita a terceeira nota: '))

media = (nota1 + nota2 + nota3) / 3 
if media >=6: 
    print(f'Aluno aprovado media: {media:.2f}')
else: 
    print(f'Aluno reprovado media: {media:.2f}')
    
```

```python
#Peça um número ao usuário e mostre a tabuada dele de 1 a 10.
num = int(input('Digite um número: '))
print(f'Tabuada do número: {num}')
for i in range(1,11):
    print(f'{num} x {i} = {num * i}')
```

```python
# Passo 1: ler a letra
letra = input('Digite uma letra: ')
letra = letra.lower( )  # garante que esteja em minúscula

# Passo 2: vogais
vogais = 'aeiou'

# Passo 3: verificar se é vogal ou consoante
if letra in vogais:
    print('A letra é uma vogal.')
else:
    print('A letra é uma consoante.')

```

```python
lado1 = float(input('Digite o primeiro lado: '))
lado2 = float(input('Digite o segundo lado: '))
lado3 = float(input('Digite o terceiro lado: '))

if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    if lado1 == lado2 != lado3:
        print('Triângulo Equilátero')
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print('Triângulo Isósceles')
    else:
        print('Triângulo Escaleno')
else:
    print('Não forma um triângulo')
```
