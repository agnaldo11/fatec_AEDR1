```python
# Calcular a média das cinco idades de alguns amigos: 
soma = 0 
for i in range(1,6):
    idade = int(input(f'Digite a idade {i}: '))
    soma += idade 
media = round(soma / i)
print(f'A soma da idades é: {soma}, e a média é: {media}')
```

```python
# Peça 3 números ao usuário e mostre:
soma = 0 
for i in range(1,4):
    num = float(input(f'Entre com m número {i}: '))
    soma += num
media = soma / 3 
print(f'A soma dos três numeros é: {soma}, e a média é: {media:.2f}')
```

```python
# Peça 4 notas de um aluno e mostre:
soma = 0 
for i in range(1,5):
    nota = int(input(f'Entre com as notas, {i}: '))
    soma += nota 
media = nota / i 
print(f'Soma das notas é: {soma}, e a média das notas {media}')
    
```

```python
# Calcular a média das cinco idades de alguns amigos: 
soma = 0 
i = 1 
while i <= 5: 
    idade = int(input(f'Entre com a idade: {i}'))
    soma += idade 
    i = i + 1 
media = round(soma / i) 
print(f'Soma das idades: {soma}, e media é: {media}')
```

```python
# calcualr n fatorial de um numero:
num = int(input('Calcule fatorial de um número: '))
fatorial = 1 
if num <= 0: 
    print('Fatorial não existe')
elif num == 0:
    print('Fatorial é igual a 1') 
else:
    for i in range(1, num + 1):
        fatorial = fatorial * i 
    print(f'Fatorial de um número: {num}, igual {fatorial}')
```
