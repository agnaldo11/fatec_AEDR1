```python
# Condicional If (Se)
if 5 < 2:
    print('Sentaça verdadeira')
else: 
    print('Sentença falça')
```

```python
# Condicional If...Else com variável
dia = 'terça'
if dia == 'segunda': 
    print('Hoje vai chover')
else:
    print('Vai ser sol')
```

```python
# Podemos usar o operador elif para validar mais de uma condição
dia = input('Digite um dia da semana:')
if dia == 'segunda':
    print('Vai chover')
elif dia == 'terça':
    print('Vai ser sol')
else: 
    print('Sem previsão do tempo')
```
