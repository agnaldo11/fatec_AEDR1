```python
lista1 = [1,2,3,4,5,6,7] 
lista2 = [3,4,5,6,7,9]
soma = 0 
for lista in [lista1, lista2]:
    for num in lista: 
        if num % 2 == 0:
            soma += num
print(f'Soma das duas lista1 + lista2 : {soma}')
```

```python
lista1 = [1,3,5,6,7,8]
lista2 = [2,3,4,5,6,7]
soma = 0 
for num in lista1 + lista2:
    if num % 2 == 0:
        soma += num 
print(f'Soma dos numeros pares: {soma}')
    
```
