```python
def addNum(n1,n2):
    print('Primeiro número: ' + str(n1))
    print('Segundo número: ' + str(n2))
    print(f'soma é:', {n1 + n2})
```

```python
from math import sqrt 
a = float(input('Digite um número para a: '))
if a == 0: 
    print('A equação não é do segundo grau')
    exit( )
b = float(input('Digite um número para b: '))
c = float(input('Digite um numero para c: '))
delta = b ** 2 - 4 * a * c 
if delta < 0: 
    print('A equação não possui raiz real')
    exit( ) 
elif delta == 0: 
    x = -b / (2 * a)
    print('A equação possui uma raiz real:', x)
else:    
    x1 = (-b + sqrt(delta)) / (2 * a)
    x2 = (-b - sqrt(delta)) / (2 * a)
    print(f'A equação possui duas raízes reais diferentes: x1 = {x1} e x2 = {x2}') 
   
```

```python
#Executar senha aleatoria: 
import random
import string 
def gerar_senha(tamanho, usar_letras=True, usar_numeros=True, usar_simbolos=True):
    caracteres = ''
    if usar_letras:
        caracteres += string.ascii_letters
    if usar_numeros:
        caracteres += string.ascii_letters
    if usar_simbolos: 
        caracteres += string.ascii_letters
    if caracteres == '':
        return('Nenhum tipo de caracteres foi selecionado')
senha = ''
tamanho = int(input('Digite sua senha: '))
for _ in range(tamanho):
    senha += random.choice(caracteres)
senha = gerar_senha(tamanho)
print('Senha gerada') 
            
    
    
```
